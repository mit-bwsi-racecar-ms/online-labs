# Path Selection 

## Overview
<p style='font-size:1.75rem;line-height:1.5'>
    The final challenge involves both line following and obstacle detection. Teams that complete the final challenge race track in the shortest amount of time wins. Shortcuts within the race track will be provided. 
    </p>

<p style='font-size:1.75rem;line-height:1.5'>
    There will be more than one possible line color to follow. Although it is possible to finish the course by just following one line, shortcuts are available if your car is able to follow more than one line. It is to your team's advantage to take these shortcuts!
    </p>

<p style='font-size:1.75rem;line-height:1.5'>
    In this section, we will be adding to our Line Following code so that it is able to detect, prioritize, and follow more than one line color!
    </p>

## Getting Started
<p style='font-size:1.75rem;line-height:1.5'>
    Here are some steps to get you started on writing your very own path selection algorithm!!
    <ol style='font-size:1.75rem;line-height:1.5'>
        <li>In <b>Step 3: Find HSV Range of the Line</b>, find the HSV threshold values for the second color. 
            <br> Save as new variables: <code>HSV_LOWER_&lt;color&gt;</code> and <code>HSV_UPPER_&lt;color&gt;</code></li>
        <br>
        <li> In the <code>line_following</code> function:
            <ul style='font-size:1.75rem;line-height:1.5'>
                <li> Create a new mask using the new HSV threshold values. Save as a new variable: <code>mask2</code>.</li>
                <li> Find new contours for the new mask. Save as a new variable: <code>contours2</code>.</li>
                <li> Call <code>get_contour_center</code> for the new contours.  Save as a new variable: <code>M2</code>.</li>
                <li> Below TASK 4, set <code>M = None</code> and <code>contours = None</code> </li>
                <li> If the first M is not None, then set <code>M</code> to the values of the first M, and set <code>contours</code> to the first contours found. 
                    <br>Else if the second M is not None, then set <code>M</code> to the values of M2 and <code>contours</code> to contours2. </li>
                <li> Note: Depending on which M you check first, that will be the color you are prioritizing.</li>
            </ul></li>
          <br>
          <li> Test your code (on a block). </li>
    </ol>
    </p>
